# Seattle Crime Data. Keeping your bike safe

```python
import pandas as pd
from bokeh.charts import Scatter, output_notebook, show, Bar
from bokeh.models import ColumnDataSource, Range1d, FactorRange
from bokeh.io import push_notebook
from bokeh.plotting import reset_output
from bokeh.plotting import figure


from datetime import datetime
from ipywidgets import interact
import colorsys
import array

output_notebook()

df =  pd.DataFrame()

seattleData = df.from_csv('./seattle_incidents_summer_2014.csv')
seattleData = seattleData.sort_values('Occurred Date or Date Range Start', 0, True, False, 'mergesort')
#seattleData = seattleData.drop(seattleData[seattleData.Longitude == 0.0].index)
#print seattleData.head

#print set(seattleData['Offense Type'])

bikeThefts = seattleData[seattleData['Offense Type'] == 'THEFT-BICYCLE']
```

In the period being measured, there were:

```python
print len(bikeThefts)
```

bicycle thefts recorded.

```python

numberBikeTheftsPerDayOfWeek = [len(bikeThefts['Occurred Date or Date Range Start'][pd.DatetimeIndex(bikeThefts['Occurred Date or Date Range Start']).dayofweek == day]) for day in range(7)] 

daysOfWeek = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

bikeTheftsDataFrame = pd.DataFrame()
bikeTheftsDataFrame['theftsPerWeekDay'] = numberBikeTheftsPerDayOfWeek
bikeTheftsDataFrame['days'] = daysOfWeek
```

```python
BikeTheftsPerDay = Bar(bikeTheftsDataFrame, label='days', values='theftsPerWeekDay', title='Number of bike thefts over the data period for each day of the week',color='Red')
BikeTheftsPerDay.x_range = FactorRange(factors=daysOfWeek)
```

## You are most likely to have your bicycle stolen on a Monday

```python
show(BikeTheftsPerDay)
```

```python
numberBikeTheftsPerHour = [len(bikeThefts['Occurred Date or Date Range Start'][pd.DatetimeIndex(bikeThefts['Occurred Date or Date Range Start']).hour == time]) for time in range(24)] 

hours = range(24)

bikeTheftsPerHourDataFrame = pd.DataFrame()
bikeTheftsPerHourDataFrame['theftsPerHour'] = numberBikeTheftsPerHour
bikeTheftsPerHourDataFrame['hours'] = hours


bikeTheftsPerHourFigure = Bar(bikeTheftsPerHourDataFrame, label='hours', values='theftsPerHour', title='Summed number of bike thefts at each hour over the whole period', color='blue')
```

## Your bicycle is most likely to be stolen at around 5pm

```python
show(bikeTheftsPerHourFigure)
```

```python
sectorNames = set(seattleData['District/Sector'])
sectorNames.pop()

bikeTheftsPerSector = [len(bikeThefts[bikeThefts['District/Sector'] == sectorName]) for sectorName in sectorNames]
print bikeTheftsPerSector

bikeTheftsPerSectorDataFrame = pd.DataFrame()
bikeTheftsPerSectorDataFrame['bikeTheftsPerSector'] = bikeTheftsPerSector
bikeTheftsPerSectorDataFrame['Sector'] = sectorNames
bikeTheftsPerSectorDataFrame = bikeTheftsPerSectorDataFrame.sort('bikeTheftsPerSector')

bikeTheftsPerSectorFigure = Bar(bikeTheftsPerSectorDataFrame, label='Sector', values='bikeTheftsPerSector', title='Number of bike thefts per sector over the date period', color='green')
bikeTheftsPerSectorFigure.x_range = FactorRange(factors=bikeTheftsPerSectorDataFrame['Sector'].tolist())
```

## Your bike is most likely to be stolen in Sector B

```python
show(bikeTheftsPerSectorFigure)
```
